```python
class Person:
  def __init__(self,PIR,DOOR_1,DOOR_2):
    self.PIR=PIR
    self.DOOR_1=DOOR_1
    self.DOOR_2=DOOR_2
  def display(self):
      print("PIR Value: ",self.PIR)
      print("Door Angle: ",self.DOOR_1)
      print("Window Angle: ",self.DOOR_2)
      if (self.PIR == 3 and self.DOOR_1 == 0 and self.DOOR_2 == 0):   
        print("Person Detected but door and window close ")
      elif(self.PIR == 3 and self.DOOR_1 > 0 and self.DOOR_2 > 0):  
        print("Person Detected with door and window open")
      elif(self.PIR == 0 and self.DOOR_1 == 0 and self.DOOR_2 == 0):   
        print("Person not Detected with door and window closed")
      elif(self.PIR == 0 and self.DOOR_1 >0 and self.DOOR_2 > 0):      
        print("Person not detected but door and window open")
      elif(self.PIR == 3 and self.DOOR_1 > 0 and self.DOOR_2 == 0):    
        print("Person Detected with door open and window closed")
      elif(self.PIR == 0 and self.DOOR_1 == 0 and self.DOOR_2 > 0):    
        print("Person not Detected with door close and window open")   
      elif(self.PIR == 0 and self.DOOR_1 > 0 and self.DOOR_2 == 0):    
        print("Person not Detected with door open and window closed")
      elif(self.PIR == 3 and self.DOOR_1 == 0 and self.DOOR_2 > 0):    
        print("Person Detected with door close and window open")
      else:
        print("PIR sensor not working")

class Moisture:
  def __init__(self,SOIL_M,GPSLA_M,GPSLO_M):
    self.SOIL_M=SOIL_M
    self.GPSLA_M=GPSLA_M
    self.GPSLO_M=GPSLO_M
  def display(self):
    print()
    print()
    print("Latitude of sensor: ",self.GPSLA_M)
    print("Longitude of sensor: ",self.GPSLO_M)
    if self.GPSLA_M>0 and self.GPSLO_M>0:
      print("Sensor is Soil Moisture")
      print("Soil Moisture Level: ",self.SOIL_M)
      if (self.SOIL_M > 500):
        print("Motor turned OFF")
      elif(self.SOIL_M <= 500 and self.SOIL_M != 0):
        print("Motor turned ON")
      else:
        print("Soil Moisture sensor not working")

class Light:
  def __init__(self,LIGHT,GPSLA_L,GPSLO_L):
    self.LIGHT=LIGHT
    self.GPSLA_L=GPSLA_L
    self.GPSLO_L=GPSLO_L
  def display(self):
    print()
    print()
    print("Latitude of sensor: ",self.GPSLA_L)
    print("Longitude of sensor: ",self.GPSLO_L)
    if self.GPSLA_L>0 and self.GPSLO_L>0:
      print("Sensor is Light")
      print("Light Value: ",self.LIGHT)
      if self.LIGHT > 900 :
        print("Intensity High")
      elif (self.LIGHT <= 900 and self.LIGHT != 0) :
        print("Intensity Low")
      else:
        print("Light sensor not working")
      
class PH:
  def __init__(self,SOIL_PH,GPSLA_P,GPSLO_P):
    self.SOIL_PH=SOIL_PH
    self.GPSLA_P=GPSLA_P
    self.GPSLO_P=GPSLO_P
  def display(self):
    print()
    print()
    print("Latitude of sensor: ",self.GPSLA_P)
    print("Longitude of sensor: ",self.GPSLO_P)
    if self.GPSLA_P>0 and self.GPSLO_P>0:
      print("Sensor is Soil pH")
      print("pH Value: ",self.SOIL_PH)
      if self.SOIL_PH > 6.7 :
        print("Too Alkaline for Plants")
      elif (self.SOIL_PH <= 5.5 and self.SOIL_PH != 0) :
        print("Too Acidic for Plants")
      elif(self.SOIL_PH > 5.5 and self.SOIL_PH <= 6.7 ):
        print("Neutral soil")
      else:
        print("Soil pH sensor not working")

class Humidity:
  def __init__(self,HUM,GPSLA_H,GPSLO_H):
    self.HUM=HUM
    self.GPSLA_H=GPSLA_H
    self.GPSLO_H=GPSLO_H
  def display(self):
    print()
    print()
    print("Latitude of sensor: ",self.GPSLA_H)
    print("Longitude of sensor: ",self.GPSLO_H)
    if self.GPSLA_H>0 and self.GPSLO_H>0:
      print("Sensor is Humidity")
      print("Humidity Value: ",self.HUM)
      if self.HUM > 75 :
        print("Humidity High")
      elif (self.HUM <= 65 and self.HUM != 0) :
        print("Humidity Low")
      elif(self.HUM > 65 and self.HUM <= 75 ):
        print("Perfect Humidity")
      else:
        print("Humidity sensor not working")

class Temperature:
  def __init__(self,TEMP,GPSLA_T,GPSLO_T):
    self.TEMP=TEMP
    self.GPSLA_T=GPSLA_T
    self.GPSLO_T=GPSLO_T
  def display(self):
    print()
    print()
    print("Latitude of sensor: ",self.GPSLA_T)
    print("Longitude of sensor: ",self.GPSLO_T)
    if self.GPSLA_T>0 and self.GPSLO_T>0:
      print("Sensor is Temperature")
      print("Temperature Value: ",self.TEMP)
      if self.TEMP > 80 :
        print("Temperature High")
      elif (self.TEMP <= 70 and self.TEMP != 0) :
        print("Temperature Low")
      elif(self.TEMP > 70 and self.TEMP <= 80 ):
        print("Perfect Temperature")
      else:
        print("Temperature sensor not working")

class Airquality:
  def __init__(self,AIR_QUAL,GPSLA_A,GPSLO_A):
    self.AIR_QUAL=AIR_QUAL
    self.GPSLA_A=GPSLA_A
    self.GPSLO_A=GPSLO_A
  def display(self):
    print()
    print()
    print("Latitude of sensor: ",self.GPSLA_A)
    print("Longitude of sensor: ",self.GPSLO_A)
    if self.GPSLA_A>0 and self.GPSLO_A>0:
      print("Sensor is Air Quality")
      print("Air Quality: ",self.AIR_QUAL)
      if self.AIR_QUAL > 600 :
        print("Polluted Air")
      elif (self.AIR_QUAL <= 600 and self.AIR_QUAL != 0) :
        print("Perfect Air")
      else:
        print("Air Quality sensor not working")
          
```

```python
from IPython.display import Javascript
display(Javascript('''google.colab.output.setIframeHeight(0, true, {maxHeight: 5000})'''))
class Context:
  def __init__(self,P):
    self.Person_=Person(df['PIR'][P],df['DOOR_1'][P],df['DOOR_2'][P])
    self.Moisture_=Moisture(df['SOIL_M'][P],df['GPSLA_M'][P],df['GPSLO_M'][P])
    self.Light_=Light(df['LIGHT'][P],df['GPSLA_L'][P],df['GPSLO_L'][P])
    self.PH_=PH(df['SOIL_PH'][P],df['GPSLA_P'][P],df['GPSLO_P'][P])
    self.Humidity_=Humidity(df['HUM'][P],df['GPSLA_H'][P],df['GPSLO_H'][P])
    self.Temperature_=Temperature(df['TEMP'][P],df['GPSLA_T'][P],df['GPSLO_T'][P])
    self.Airquality_=Airquality(df['AIR_QUAL'][P],df['GPSLA_A'][P],df['GPSLO_A'][P])

  def display(self):
    self.Person_.display()
    self.Moisture_.display()
    self.Light_.display()
    self.PH_.display()
    self.Humidity_.display()
    self.Temperature_.display()
    self.Airquality_.display()


if __name__=="__main__":
  counter=1
  contextList=[]
  for i in range(50):
    context = Context(i)
    contextList.append(context)
  for context in contextList:
    print("------------------Context Behaviour---------------------")
    print()
    context.display()
    counter=counter+1
    print()
    print()
```
